## Check for missing values

### Subtask:
Identify and handle any missing values in the dataset.


**Reasoning**:
Check for and handle missing values by dropping rows with any missing values, then verify the removal.



```python
missing_values = df.isnull().sum()
print("Missing values before dropping:")
display(missing_values)

df.dropna(inplace=True)

missing_values_after_drop = df.isnull().sum()
print("\nMissing values after dropping:")
display(missing_values_after_drop)
```

## Analyze categorical features

### Subtask:
Explore the distribution of categorical variables like 'Party' and 'State_Name'.


**Reasoning**:
Calculate and display the value counts for the 'Party' and 'State_Name' columns to explore their distributions.



```python
top_n = 15
top_parties = df['Party'].value_counts().head(top_n)
print(f"Top {top_n} most frequent parties:")
display(top_parties)

top_states = df['State_Name'].value_counts().head(top_n)
print(f"\nTop {top_n} most frequent states:")
display(top_states)
```

## Analyze numerical features

### Subtask:
Explore the distribution of numerical variables like 'Votes', 'Valid_Votes', and 'Vote_Share'.


**Reasoning**:
Calculate and display descriptive statistics for the numerical columns and then calculate and display the correlation matrix. Finally, extract and display specific correlation values.



```python
print("Descriptive statistics for numerical columns:")
display(df[['Votes', 'Valid_Votes', 'Vote_Share']].describe())

numerical_cols = ['Votes', 'Valid_Votes', 'Vote_Share', 'Constituency_No', 'Year', 'N_Cand', 'Incumbent']
correlation_matrix = df[numerical_cols].corr()
print("\nCorrelation matrix for numerical columns:")
display(correlation_matrix)

corr_votes_share = correlation_matrix.loc['Votes', 'Vote_Share']
corr_votes_valid = correlation_matrix.loc['Votes', 'Valid_Votes']
corr_valid_share = correlation_matrix.loc['Valid_Votes', 'Vote_Share']

print(f"\nCorrelation between Votes and Vote_Share: {corr_votes_share}")
print(f"Correlation between Votes and Valid_Votes: {corr_votes_valid}")
print(f"Correlation between Valid_Votes and Vote_Share: {corr_valid_share}")
```

## Analyze time series data

### Subtask:
Analyze the distribution of 'Year' and how other variables change over time.


**Reasoning**:
Calculate and display the value counts for the 'Year' column and then group the DataFrame by 'Year' and calculate the mean of specified numerical columns.



```python
year_counts = df['Year'].value_counts()
print("Distribution of data across election years:")
display(year_counts)

yearly_means = df.groupby('Year')[['Votes', 'Valid_Votes', 'Vote_Share', 'N_Cand', 'Incumbent']].mean()
print("\nYearly means of numerical variables:")
display(yearly_means)
```

## Visualize data

### Subtask:
Create visualizations to better understand the data and its patterns.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from datetime import datetime
from pathlib import Path

# Go up one level from notebooks folder to project root
output_path = Path('../reports')

# Create parent directories if they don't exist
output_path.mkdir(parents=True, exist_ok=True)

# Prepare data within this cell so it runs independently
_top_parties = df['Party'].value_counts().head(10).index
df_top_parties = df[df['Party'].isin(_top_parties)]
_num_cols = ['Votes', 'Valid_Votes', 'Vote_Share', 'Constituency_No', 'Year', 'N_Cand']
correlation_matrix = df[_num_cols].corr()

# Set style
sns.set_style("whitegrid")
plt.rcParams['font.size'] = 10

# Create a 2x2 grid with enhanced styling
fig, axes = plt.subplots(2, 2, figsize=(18, 14))
fig.suptitle('Comprehensive Election Data Analysis',
            fontsize=20, fontweight='bold', y=0.95)

# Define colors
colors = ['#2E86AB', '#A23B72', '#F18F01', '#C73E1D']

# 1. Enhanced Histogram of Vote Share
sns.histplot(df['Vote_Share'], kde=True, ax=axes[0, 0],
             color=colors[0], alpha=0.7, edgecolor='black', linewidth=0.5)
axes[0, 0].set_title('A) Distribution of Vote Share', fontweight='bold', pad=15)
axes[0, 0].set_xlabel('Vote Share (%)')
axes[0, 0].set_ylabel('Frequency')

# Add comprehensive statistics
stats_text = f'Mean: {df["Vote_Share"].mean():.1f}%\nMedian: {df["Vote_Share"].median():.1f}%\nStd: {df["Vote_Share"].std():.1f}%'
axes[0, 0].text(0.95, 0.95, stats_text, transform=axes[0, 0].transAxes,
               verticalalignment='top', horizontalalignment='right',
               bbox=dict(boxstyle='round', facecolor='white', alpha=0.8),
               fontweight='bold')

# 2. Enhanced Boxplot with colors
sns.boxplot(x='Party', y='Vote_Share', data=df_top_parties, ax=axes[0, 1], palette='Set3')
axes[0, 1].set_title('B) Vote Share by Top 10 Parties', fontweight='bold', pad=15)
axes[0, 1].set_xlabel('Party')
axes[0, 1].set_ylabel('Vote Share (%)')
axes[0, 1].tick_params(axis='x', rotation=45)

# Add mean points
party_means = df_top_parties.groupby('Party')['Vote_Share'].mean().values
for i, mean_val in enumerate(party_means):
    axes[0, 1].plot(i, mean_val, 'o', color='red', markersize=6, label='Mean' if i == 0 else "")

# 3. Enhanced Scatter plot with regression line
sns.regplot(x='Valid_Votes', y='Votes', data=df,
           scatter_kws={'alpha':0.5, 'color': colors[2]},
           line_kws={'color': 'red', 'linewidth': 2},
           ax=axes[1, 0])
axes[1, 0].set_title('C) Votes vs Valid Votes', fontweight='bold', pad=15)
axes[1, 0].set_xlabel('Valid Votes')
axes[1, 0].set_ylabel('Votes')

# 4. Enhanced Correlation Matrix
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))  # Mask upper triangle
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu_r', fmt=".2f",
           center=0, ax=axes[1, 1], mask=mask,
           cbar_kws={'shrink': 0.8})
axes[1, 1].set_title('D) Correlation Matrix', fontweight='bold', pad=15)

# Improve correlation matrix labels rotation
axes[1, 1].set_xticklabels(axes[1, 1].get_xticklabels(), rotation=45, ha='right')
axes[1, 1].set_yticklabels(axes[1, 1].get_yticklabels(), rotation=0)

plt.tight_layout()
plt.subplots_adjust(top=0.92)

# Save the plot to the reports directory
filepath = output_path / 'comprehensive_analysis.png'
plt.savefig(str(filepath), dpi=300, bbox_inches='tight', facecolor='white')
print(f"Plot saved as: {filepath}")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Go up one level from notebooks folder to project root
output_path = Path('../reports')

# Create parent directories if they don't exist
output_path.mkdir(parents=True, exist_ok=True)

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Election Data Analysis Dashboard', fontsize=16, fontweight='bold')

# 1. Histogram of Valid Votes (Top Left)
sns.histplot(df['Valid_Votes'], kde=True, ax=axes[0, 0])
axes[0, 0].set_title('A) Distribution of Valid Votes', fontweight='bold')
axes[0, 0].set_xlabel('Valid Votes')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].grid(True, alpha=0.3)

# Add statistics annotation
mean_valid = df['Valid_Votes'].mean()
median_valid = df['Valid_Votes'].median()
axes[0, 0].axvline(mean_valid, color='red', linestyle='--', alpha=0.8, label=f'Mean: {mean_valid:.0f}')
axes[0, 0].axvline(median_valid, color='green', linestyle='--', alpha=0.8, label=f'Median: {median_valid:.0f}')
axes[0, 0].legend()

# 2. Boxplot of Votes by Constituency Type (Top Right)
sns.boxplot(x='Constituency_Type', y='Votes', data=df, ax=axes[0, 1])
axes[0, 1].set_title('B) Votes Distribution by Constituency Type', fontweight='bold')
axes[0, 1].set_xlabel('Constituency Type')
axes[0, 1].set_ylabel('Votes')
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Countplot of Incumbent (Bottom Left)
sns.countplot(x='Incumbent', data=df, ax=axes[1, 0])
axes[1, 0].set_title('C) Distribution of Incumbent Status', fontweight='bold')
axes[1, 0].set_xlabel('Incumbent')
axes[1, 0].set_ylabel('Count')

# Add count labels on bars
for container in axes[1, 0].containers:
    axes[1, 0].bar_label(container, fmt='%d', padding=3)

# 4. Additional Analysis - Valid Votes by Constituency Type (Bottom Right)
sns.boxplot(x='Constituency_Type', y='Valid_Votes', data=df, ax=axes[1, 1])
axes[1, 1].set_title('D) Valid Votes by Constituency Type', fontweight='bold')
axes[1, 1].set_xlabel('Constituency Type')
axes[1, 1].set_ylabel('Valid Votes')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.subplots_adjust(top=0.93)

# Save the figure
# Save the plot to the reports directory
filepath = output_path / 'election_data_dashboard.png'
plt.savefig(str(filepath), dpi=300, bbox_inches='tight', facecolor='white')
print(f"Plot saved as: {filepath}")
plt.show()
```
