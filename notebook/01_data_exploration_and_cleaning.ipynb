## Handle missing values

### Subtask:
Address missing values in relevant columns.


**Reasoning**:
Check the number of null values in the relevant columns to identify how to handle the missing values.



```python
print(df[['Votes', 'Electors', 'Party', 'State_Name']].isnull().sum())
```

**Reasoning**:
Since 'Votes', 'Electors', and 'State_Name' have no missing values, and 'Party' has only 2 missing values, drop the rows with missing values in the 'Party' column as this is a small number and imputation might not be accurate.



```python
df.dropna(subset=['Party'], inplace=True)
print(df[['Votes', 'Electors', 'Party', 'State_Name']].isnull().sum())
```

## Standardize data

### Subtask:
Clean and standardize 'State_Name' and 'Party' columns.


**Reasoning**:
Standardize the 'State_Name' and 'Party' columns by converting them to lowercase and then display the unique values and their counts to verify the changes.



```python
df['State_Name'] = df['State_Name'].str.lower()
df['Party'] = df['Party'].str.lower()

display(df['State_Name'].value_counts())
display(df['Party'].value_counts())
```

## Select relevant columns

### Subtask:
Keep only the necessary columns for the analysis.


**Reasoning**:
Filter the DataFrame to keep only the essential columns and display the head and info of the filtered DataFrame.



```python
columns_to_keep = ['State_Name', 'Constituency_No', 'Year', 'Party', 'Candidate','Incumbent','N_Cand','Constituency_Type', 'Votes', 'Valid_Votes']
df = df[columns_to_keep]
display(df.head())
display(df.info())
```

```python
df['Incumbent'].fillna('True',inplace=True)
```

## Calculate vote share

### Subtask:
Compute the vote share percentage for each candidate.


**Reasoning**:
Calculate the vote share percentage for each candidate using the 'Votes' and 'Valid_Votes' columns, and then display the head of the DataFrame to show the new 'Vote_Share' column.



```python
df['Vote_Share'] = (df['Votes'] / df['Valid_Votes']) * 100
display(df.head())
```
